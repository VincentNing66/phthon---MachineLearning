```python
from sklearn.model_selection import train_test_split
X = df.iloc[:, 1:-1]
y = df.iloc[:, 9]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1345375)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
op_obj_df = df.select_dtypes(include=['object']).copy()
print(op_obj_df['ocean_proximity'].value_counts())
```

```python
replace_map = {'ocean_proximity': {'NEAR BAY': 1, '<1H OCEAN': 2, 'NEAR OCEAN': 3, 'ISLAND': 4,
                                  'INLAND':5}}
labels = op_obj_df['ocean_proximity'].astype('category').cat.categories.tolist()
replace_map_comp = {'ocean_proximity' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}

op_obj_df_replace = op_obj_df.copy()
op_obj_df_replace.replace(replace_map_comp, inplace=True)

X_train['ocean_proximity'] = op_obj_df_replace['ocean_proximity']
X_test['ocean_proximity'] = op_obj_df_replace['ocean_proximity']
X_train.dtypes.sample(8)
#X_test.dtypes.sample(8)
```

```python
def run_reg( regressor, X_train, X_test, y_train, y_test):
    from sklearn.metrics import mean_squared_error
    import matplotlib.pyplot as plt
    my_ridge = regressor
    my_ridge.fit(X_train, y_train)
    
    pred = my_ridge.predict(X_test)
    pred[ pred < 15000 ] = 15000
    pred[ pred > 500000 ] = 500000
    
    mse = mean_squared_error(y_test, pred)
    
    plt.title('MAE = ' + str(np.sqrt(mse)))
    plt.scatter(pred,y_test)
    plt.xlabel('Prediction')
    plt.ylabel('Test')
    plt.show()
    plt.plot(my_ridge.coef_)
    plt.show()
    return np.sqrt(mse)
```

```python
from sklearn import linear_model
R1 = linear_model.RidgeCV(alphas=[0.1])
run_reg(R1,X_train,X_test,y_train,y_test)
```

```python
R2 = linear_model.RidgeCV(alphas=[0.001])
run_reg(R2,X_train,X_test,y_train,y_test)
```

```python
from sklearn.linear_model import Lasso
L1 = Lasso(alpha = 0.1)
run_reg(L1,X_train,X_test,y_train,y_test)
```

```python
L2 = Lasso(alpha = 0.001)
run_reg(L2,X_train,X_test,y_train,y_test)
```

```python
from sklearn.linear_model import ElasticNet
E1 = ElasticNet(alpha=0.1)
run_reg(E1,X_train,X_test,y_train,y_test)
```

```python
E2 = ElasticNet(alpha=0.001)
run_reg(E2,X_train,X_test,y_train,y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=3, include_bias=False)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)
```

```python
R3 = linear_model.RidgeCV(alphas=[0.1])
run_reg(R3,X_train_poly,X_test_poly,y_train,y_test)
```
