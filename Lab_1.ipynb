```python
# load Data
loadData = pd.read_csv('../input/wisconsin_breast_cancer.csv')
loadData = loadData.fillna(0)
loadData.head()
```

```python
loadData.info(0)
```

```python
# Visualising it, using a pairplot
import seaborn as sns
sns.pairplot(data = loadData, hue = 'class', palette = 'Set1')
```

```python
#Selecting all features except for "ID" and "class" as X,
#  selecting "class" as y, and splitting into train and test
#  with test_size=0.2, or 20%
from sklearn.model_selection import train_test_split
x = loadData.iloc[:, 2:-2]
# generate an xsub_train from x_train,
# and an xsub_test from x_test
xSub = loadData.iloc[:, 3:7]
y = loadData.iloc[:, -1]
ySub = loadData.iloc[:, -1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
xsub_train, xsub_test, ysub_train, ysub_test = train_test_split(xSub, ySub, test_size = 0.2)
print(x_train.shape, y_train.shape)
print(xsub_train.shape, ysub_train.shape)
```

```python
# Train an SVC classifier on the train set
from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
#Train another SVC classifier on xsub_train and y_train,
modelSub = SVC()
modelSub.fit(xsub_train,ysub_train)
```

```python
# predict for the test set
pred = model.predict(x_test)
# predict for xsub_test
predSub = modelSub.predict(xsub_test)
print(pred[:-1])
print(y_test[:-1])
print(predSub[:-1])
print(y_test[:-1])
```

```python
# generates the confusion_matrix
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, pred))
print(confusion_matrix(ysub_test, predSub))
```

```python
# generates classification_report
print(classification_report(y_test, pred))
print(classification_report(ysub_test, predSub))
# by reading the data y sub one's prediction is more accaurate then the firast one
# obviously the second pair of data with higher difference between data has more potential pridiction
```

```python
from sklearn.neighbors import KNeighborsClassifier

#Create a KNN with Classifier = 5
knn = KNeighborsClassifier(n_neighbors=1)

#Train the model using the pervious training sets
knn.fit(x_train, y_train)

#Predict the response for test dataset
y_pred2 = knn.predict(x_test)
print(y_pred2)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred2))
```

```python
#Create a KNN with Classifier = 5
knn = KNeighborsClassifier(n_neighbors=5)

#Train the model using the training sets
knn.fit(x_train, y_train)

#Predict the response for test dataset
y_pred3 = knn.predict(x_test)
print(y_pred3)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred3))
```

```python
# generates classification_report for knn = 1
print(confusion_matrix(y_test, y_pred2))
# generates classification_report for knn = 5
print(confusion_matrix(y_test, y_pred3))
```

```python
# generates classification_report for knn = 1
print(classification_report(y_test, y_pred2))
# generates classification_report for knn = 5
print(classification_report(y_test, y_pred3))
#Obviously , the data with knn = 5 has more accuracy 
# That means knn with larger neighbors has more accuracy
```
