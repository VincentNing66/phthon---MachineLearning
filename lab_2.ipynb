```python
# load Data
loadData = pd.read_csv('../input/wisconsin_breast_cancer.csv')
loadData = loadData.fillna(0)
loadData.head()
```

```python
loadData.info()
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import train_test_split
x = loadData.iloc[:, 1:-1]
y = loadData.iloc[:, 10]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=1345375)
print(x_train.shape, y_train.shape)
```

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=1345375)
sgd_clf.fit(x_train,y_train)
```

```python
columns = [ 'thickness', 'size', 'shape', 'adhesion', 'single', 'nuclei', 'chromatin', 'nucleoli', 'mitosis' ]
columnID = [0, 0, 0, 0, 0, 0, 0, 0, 0]
current =[]
```

```python
from sklearn.model_selection import cross_val_score
finalScores = []
finalACs = []
xsub = loadData[[]]
from sklearn.metrics import accuracy_score
```

```python
for i in range(1,512):
    current =[]
    for j in range(2,len(str(bin(i)))):
        columnID[j-2] = 0
        if str(bin(i))[j]=='1':
            columnID[j-2]=1
            current.append(columns[j-2])
    xsub = x_train[current]
    scores = cross_val_score(sgd_clf,xsub,y_train,scoring='accuracy',cv=10)
    finalScores.append(np.mean(scores))
    sgd_clf.fit(xsub,y_train)
    
    preds = sgd_clf.predict(x_test[current])
    ac = accuracy_score(y_test,preds)
    finalACs.append(ac)
```

```python
maxNum = 0
maxIndex = 0
maxNumAC = 0
maxIndexAC = 0
for i in range(511):
    if finalScores[i] > maxNum:
        maxNum = finalScores[i]
        maxIndex = i
for i in range(511):
    if finalACs[i] > maxNumAC:
        maxNumAC = finalACs[i]
        maxIndexAC = i

print(maxNum, maxIndex, bin(maxIndex), maxNumAC, maxIndexAC, bin(maxIndexAC))
```

```python
import matplotlib.pyplot as plt
plt.scatter(finalScores, finalACs)
plt.title('comparision between cv accuracy and test accuracy')
plt.xlabel('CV Accuracy')
plt.ylabel('Test Accuracy')
plt.show()
```
